```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_prices = pd.read_csv('stocks_daily_prices-230525-093933.csv')
df_prices
```

```python
df_returns = pd.read_csv('stocks_daily_returns-230525-093933.csv')
df_returns
```

```python
df_prices['Date'] = pd.to_datetime(df_prices['Date'])
df_returns['Date'] = pd.to_datetime(df_returns['Date'])
```

```python
df_prices.plot(x='Date', y=['AAPL','JPM','PG','UAL'], linewidth=3, figsize=(16,8))
plt.ylabel('Price')
plt.title('Stock Prices')
plt.legend(loc = 'upper center')
plt.grid();
```

```python
df_returns.plot(x='Date', y=['AAPL','JPM','PG','UAL'], subplots=True, grid=True, linewidth=3, figsize=(16,16))
plt.ylabel('Price')
plt.title('Stock Prices')
plt.legend(loc = 'upper center');
```

```python
plt.figure(figsize=(20, 4))

plt.subplot(1, 4, 1)
plt.plot(df_returns['AAPL'], 'r--')
plt.grid();

plt.subplot(1, 4, 2)
plt.plot(df_returns['JPM'], 'b.')
plt.grid();

plt.subplot(1, 4, 3)
plt.plot(df_returns['PG'], 'g--')
plt.grid();

plt.subplot(1, 4, 4)
plt.plot(df_returns['UAL'], 'y.')
plt.grid();
```

```python
X = df_returns['AAPL']
Y = df_returns['JPM']
plt.figure(figsize=(15,10))
plt.scatter(X, Y)
plt.grid();
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='AAPL', y='JPM', data=df_returns)
plt.grid();
```

```python
values = [10, 70, 10, 10]
colors = ['g', 'r', 'y', 'b']
explode = [0, 0.2, 0, 0]
labels = ['JP Morgan Chase (JPM)', 'APPLE (AAPL)', 'United Airlines (UAL)', 'P&G (PG)']

plt.figure(figsize=(10, 10))
plt.pie(values, colors=colors, labels=labels, explode=explode)
plt.title('Sport Portfolio')
plt.show();
```

```python
plt.figure(figsize=(20,15));

mu_UAL = round(df_returns['UAL'].mean(), 2)
sigma_UAL = round(df_returns['UAL'].std(), 2)

mu_PG = round(df_returns['PG'].mean(), 2)
sigma_PG = round(df_returns['PG'].std(), 2)

df_returns[['PG', 'UAL']].plot.hist(bins=60, alpha=0.7, figsize=(15,8))
plt.grid();
plt.ylabel('Probability');

plt.title(f'Histogram: P&G mu={mu_PG}, P&G sigma={sigma_PG}, United Airlines mu={mu_UAL}, United Airlines sigma={sigma_UAL}');
plt.show();
```

```python
plt.figure(figsize=(12,10))
cm = df_returns.corr()
sns.heatmap(cm, annot=True);
```
