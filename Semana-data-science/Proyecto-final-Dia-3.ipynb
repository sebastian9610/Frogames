```python
df_life_expectancy = pd.get_dummies(df_life_expectancy, columns=['Status'], dtype=int)
```

```python
df_life_expectancy
```

```python
null_counts = df_life_expectancy.isnull().sum()
null_counts[null_counts != 0]
```

```python
df_life_expectancy.fillna(df_life_expectancy.mean(), inplace=True)
```

```python
df_life_expectancy.info()
```

```python
X = df_life_expectancy.drop(['Life expectancy '], axis=1)
y = df_life_expectancy[['Life expectancy ']]
```

```python
print('X:', X.shape)
print('y:', y.shape)
```

```python
X = np.array(X).astype('float32')
y = np.array(y).astype('float32')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
import xgboost as xgb

model = xgb.XGBRegressor(objective='reg:squarederror', learning_rate=0.1, max_depth=30, n_estimators=100)
model.fit(X_train, y_train)
```

```python
model_accuracy = model.score(X_test, y_test)
model_accuracy
```

```python
y_predict = model.predict(X_test)
y_predict 
```

```python
plt.plot(y_test, y_predict, '^', color='g');
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

k = X_test.shape[1]
n = len(X_test)

RMSE = round(float(np.sqrt(mean_squared_error(y_test, y_predict))), 3)
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2)-(n-1)/(n-k-1)

print(f'RMSE: {RMSE}\nMSE: {MSE}\nMAE: {MAE}\nr2: {r2}\nAdjusted R2:{adj_r2}')
```
