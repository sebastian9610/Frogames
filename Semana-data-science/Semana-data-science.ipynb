### tarea de programacion #3

```python
X = df_university.drop(columns=['Chance_of_Admission'])
y = df_university['Chance_of_Admission']
```

```python
X = np.array(X)
y = np.array(y)
```

```python
# Redimensionar el array de (1000,) a (1000, 1)
y = y.reshape(-1,1)
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

XGBOOST no requiere de la estandarización y normalización de los datos, lo hace automaticamente

### Tarea de programacion #4

```python
import xgboost as xgb

model = xgb.XGBRegressor(objective='reg:squarederror', learning_rate=0.1, max_depth=2, n_estimators=100)
model.fit(X_train, y_train)
```

```python
# Predecir la puntuacion del modelo entrenado utilizando el conjunto de datos de prueba

results = model.score(X_test, y_test)
print(f'Accuracy: {results}')
```

```python
y_predict = model.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

RMSE = round(float(np.sqrt(mean_squared_error(y_test, y_predict))), 3)
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)

print(f'RMSE: {RMSE}\nMSE: {MSE}\nMAE: {MAE}\nr2: {r2}')
```
