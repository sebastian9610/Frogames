### tarea de programacion #7: Funciones con Pandas

```python
# Supongamos que el dayly rate ha aumentado un 10%
# Definimos una funcion que aplique los cambios en la columna daily rate
def daily_rate_update(balance):
    return balance * 1.1
```

```python
df_hr[['DailyRate']] 
```

```python
df_hr['DailyRate'] = df_hr['DailyRate'].apply(daily_rate_update)
df_hr[['DailyRate']]
```
