## 3. Preparar los datos del modelo y dividirlos en entrenamiento y test

Como hay variables categoricas, se debe hacer un one hot encoded

```python
from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder()
df_cat = onehotencoder.fit_transform(df_UCI_Credit_Card[['SEX','EDUCATION','MARRIAGE']].copy()).toarray()
df_num = df_UCI_Credit_Card.select_dtypes(include='number').copy()
df_num.drop(['SEX','EDUCATION','MARRIAGE'], axis=1, inplace=True)
```

```python
df_cat = pd.DataFrame(df_cat)
```

```python
X = pd.concat([df_num, df_cat], axis=1)
X
```

```python
X.drop(['default.payment.next.month'], axis=1, inplace=True)
```

```python
y = df_UCI_Credit_Card[['default.payment.next.month']]
```

```python
X.columns = X.columns.astype(str)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X = scaler.fit_transform(X.astype(float))
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

#X = np.array(df_UCI_Credit_Card.drop(['default.payment.next.month'], axis=1))
#y = np.array(df_UCI_Credit_Card[['default.payment.next.month']])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape
```

## 4. Entrenar y evaluar un modelo de clasificador XG-Boost

```python
import xgboost as xgb

model_xgb = xgb.XGBClassifier(learning_rate=0.1, max_depth=20, use_label_encoder=False)
model_xgb.fit(X_train, y_train)
```

```python
results = model_xgb.score(X_test, y_test)
print(f'Precision: {results}')
```

```python
y_predict = model_xgb.predict(X_test)
y_predict
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, confusion_matrix, classification_report, accuracy_score

RMSE = round(float(np.sqrt(mean_squared_error(y_test, y_predict))), 3)
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)

print(f'RMSE: {RMSE}\nMSE: {MSE}\nMAE: {MAE}\nr2: {r2}')
```

```python
print(f'Accuracy: {100*accuracy_score(y_predict, y_test)} %')
```

```python
sns.heatmap(confusion_matrix(y_predict, y_test), annot=True, fmt='d');
```

```python
print(classification_report(y_test, y_predict))
```

## 5. Entrenar y evaluar un modelo de regresion logistica

```python
from sklearn.linear_model import LogisticRegression

model_LR = LogisticRegression(max_iter=100000)
model_LR.fit(X_train, y_train)
```

## 10. Encontrar las curvas ROC 

```python
from sklearn.metrics import roc_curve

fpr0, tpr0, thresh0 = roc_curve(y_test, model_xgb.predict_proba(X_test)[:, 1], pos_label=1)
fpr1, tpr1, thresh1 = roc_curve(y_test, model_SVM.predict_proba(X_test)[:, 1], pos_label=1)
fpr2, tpr2, thresh2 = roc_curve(y_test, model_LR.predict_proba(X_test)[:, 1], pos_label=1)
fpr3, tpr3, thresh3 = roc_curve(y_test, model_rf.predict_proba(X_test)[:, 1], pos_label=1)
fpr4, tpr4, thresh4 = roc_curve(y_test, model_gnb.predict_proba(X_test)[:, 1], pos_label=1)
fpr5, tpr5, thresh5 = roc_curve(y_test, model_knn.predict_proba(X_test)[:, 1], pos_label=1)
```

```python
from sklearn.metrics import roc_auc_score

auc_score0 = roc_auc_score(y_test, model_xgb.predict_proba(X_test)[:, 1])
auc_score1 = roc_auc_score(y_test, model_SVM.predict_proba(X_test)[:, 1])
auc_score2 = roc_auc_score(y_test, model_LR.predict_proba(X_test)[:, 1])
auc_score3 = roc_auc_score(y_test, model_rf.predict_proba(X_test)[:, 1])
auc_score4 = roc_auc_score(y_test, model_gnb.predict_proba(X_test)[:, 1])
auc_score5 = roc_auc_score(y_test, model_knn.predict_proba(X_test)[:, 1])

print(f'XG-Boost Classifier: {auc_score0}')
print(f'SVM: {auc_score1}')
print(f'Regresion Logistica: {auc_score2}')
print(f'Random Forest: {auc_score3}')
print(f'Naive Bayes: {auc_score4}')
print(f'KNN: {auc_score5}')
```

```python
plt.plot(fpr0, tpr0, linestyle='--', color='blue', label='XG-Boost')
plt.plot(fpr1, tpr1, linestyle='--', color='red', label='SVM')
plt.plot(fpr2, tpr2, linestyle='--', color='green', label='LR')
plt.plot(fpr3, tpr3, linestyle='--', color='orange', label='RF')
plt.plot(fpr4, tpr4, linestyle='--', color='purple', label='NB')
plt.plot(fpr5, tpr5, linestyle='--', color='black', label='KNN')

plt.title('ROC')
plt.xlabel('Falsos Positivos')
plt.ylabel('Verdaderos Positivos')

plt.legend(loc='best')
plt.show();
```
